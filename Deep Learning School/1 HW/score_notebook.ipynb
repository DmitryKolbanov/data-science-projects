```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('game_of_thrones_train.csv', index_col='S.No')
test = pd.read_csv('game_of_thrones_test.csv', index_col='S.No')
train.head()
```

```python
train.info()
```

```python
train.drop(columns=['dateOfBirth', 'mother', 'father', 'heir', 'spouse', 'age'], inplace=True)
test.drop(columns=['dateOfBirth', 'mother', 'father', 'heir', 'spouse', 'age'], inplace=True)
```

```python
train['title'].fillna('No title', inplace=True)
train['house'].fillna('No house', inplace=True)
train['culture'].fillna('No info', inplace=True)

train['isAliveFather'].fillna(0, inplace=True)
train['isAliveMother'].fillna(0, inplace=True)
train['isAliveHeir'].fillna(0, inplace=True)
train['isAliveSpouse'].fillna(0, inplace=True)


test['title'].fillna('No title', inplace=True)
test['house'].fillna('No house', inplace=True)
test['culture'].fillna('No info', inplace=True)

test['isAliveFather'].fillna(0, inplace=True)
test['isAliveMother'].fillna(0, inplace=True)
test['isAliveHeir'].fillna(0, inplace=True)
test['isAliveSpouse'].fillna(0, inplace=True)
```

```python
train['isPopular'] = train['popularity'].apply(lambda x: 1 if x >= 0.5 else 0)
train['boolDeadRelations'] = train['numDeadRelations'].apply(lambda x: 1 if x > 0 else 0)

test['isPopular'] = test['popularity'].apply(lambda x: 1 if x >= 0.5 else 0)
test['boolDeadRelations'] = test['numDeadRelations'].apply(lambda x: 1 if x > 0 else 0)
```

```python
cult = {
    'Summer Islands': ['summer islands', 'summer islander', 'summer isles'],
    'Ghiscari': ['ghiscari', 'ghiscaricari',  'ghis'],
    'Asshai': ["asshai'i", 'asshai'],
    'Lysene': ['lysene', 'lyseni'],
    'Andal': ['andal', 'andals'],
    'Braavosi': ['braavosi', 'braavos'],
    'Dornish': ['dornishmen', 'dorne', 'dornish'],
    'Myrish': ['myr', 'myrish', 'myrmen'],
    'Westermen': ['westermen', 'westerman', 'westerlands'],
    'Westerosi': ['westeros', 'westerosi'],
    'Stormlander': ['stormlands', 'stormlander'],
    'Norvoshi': ['norvos', 'norvoshi'],
    'Northmen': ['the north', 'northmen'],
    'Free Folk': ['wildling', 'first men', 'free folk'],
    'Qartheen': ['qartheen', 'qarth'],
    'Reach': ['the reach', 'reach', 'reachmen'],
}
```

```python
cult_reversed = {}
for key, values in cult.items():
    for value in values:
        cult_reversed[value] = key
```

```python
train['culture'] = train['culture'].apply(lambda x: cult_reversed[x.lower(
)] if x.lower() in cult_reversed.keys() else x.capitalize())

test['culture'] = test['culture'].apply(lambda x: cult_reversed[x.lower(
)] if x.lower() in cult_reversed.keys() else x.capitalize())
```

```python
train.drop(columns=['name', 'boolDeadRelations', 'isPopular', 'isAliveMother',
           'isAliveFather', 'isAliveHeir', 'isAliveSpouse', 'isNoble'], inplace=True)
test.drop(columns=['name', 'boolDeadRelations', 'isPopular', 'isAliveMother',
          'isAliveFather', 'isAliveHeir', 'isAliveSpouse', 'isNoble'], inplace=True)
```

```python
train.info()
```

```python
X, y = train.drop(columns='isAlive'), train['isAlive']
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
from catboost import CatBoostClassifier

cat_cols = ['title', 'culture', 'house']

catboost = CatBoostClassifier(eval_metric='Accuracy', cat_features=cat_cols, verbose=False)

# подберем гиперпараметры
params = {'depth': np.arange(2, 8),
          'l2_leaf_reg': np.arange(0.1, 2, 0.5),
          'learning_rate': np.arange(0.01, 0.5, 0.01),
          'n_estimators': np.arange(300, 1000, 200),
          'min_data_in_leaf': np.arange(100, 1001, 100)}

grid_search = RandomizedSearchCV(catboost,
                           param_distributions=params,
                           scoring='accuracy',
                           cv=5,
                           refit=True,
                           n_jobs=-1)

grid_search.fit(X, y)
print(f'Accuracy: {grid_search.best_score_:.4f}')
print(f"Best params: {grid_search.best_params_}")
```

```python
submission = pd.read_csv("submission.csv")
```

```python
submission['isAlive'] = grid_search.predict(test)
```

```python
submission.to_csv("my_submission.csv", index=False)
```
